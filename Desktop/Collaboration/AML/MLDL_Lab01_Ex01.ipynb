### 2. Create a random tensor with shape `(7, 7)`.


````python
# Import torch
import torch
# Create random tensor
random_tensor = torch.rand(7, 7)
print(random_tensor)
print(f"Shape of the tensor: {random_tensor.shape}")
````

### 3. Perform a matrix multiplication on the tensor from 2 with another random tensor with shape `(1, 7)` (hint: you may have to transpose the second tensor).

````python
# Create another random tensor
random_tensor2 = torch.rand(1, 7)
print(random_tensor2)
print(f"Shape of the tensor: {random_tensor2.shape}")

# Perform matrix multiplication
transposed_tensor= torch.transpose(random_tensor2, 0, 1)
print(transposed_tensor)
print(f"Shape of the tensor: {transposed_tensor.shape}")

result = torch.matmul(random_tensor, transposed_tensor)
print(result)
````
